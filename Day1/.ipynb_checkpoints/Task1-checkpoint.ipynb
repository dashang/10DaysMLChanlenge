```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dt = pd.read_csv("time_series_2019-ncov-Confirmed.csv")
```

```python
dt.head()
```

## Task: 
 - Processed the data: On a particular date, If 70% number of confirmed case is zero, then Delete the column. i.e. whole February will be deleted and few more.
 - Plot the graph
     - Country Wise
     - Date Wise
     - Continent Wise 

```python
dt['Country/Region'].shape
```

```python
for column in dt[dt.columns[4:]]:
    if (dt[column].isin([0]).sum() / 487) * 100 > 70:
        dt.drop(column,axis=1, inplace=True)
```

```python
dt.head()
```

```python
dt["Country/Region"].unique() , len(dt["Country/Region"].unique())
```

```python
dt.isnull().sum()
```

## Number of Cases Country-Wise

```python
droped_state = dt.drop(['Province/State'],axis=1)
```

```python
dt_grp_Country = droped_state.groupby("Country/Region").sum()
```

```python
dt_grp_Country.head()
```

```python
dt_Country_desc= dt_grp_Country.sort_values(by='3/22/20', ascending=False)
```

```python
dt_Country_desc
```

```python
Top_Countries = dt_Country_desc.head(10)
```

```python
plt.bar(Top_Countries.index,Top_Countries["3/22/20"],align='center')
plt.title("Country-wise Cases")
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Cases')
plt.show()
```

```python
Top_Countries.iloc[0,2:].index
```

## Date Wise Cases

```python
plt.bar(Top_Countries.iloc[0,2:9].index,Top_Countries.iloc[0,2:9],align='center')
plt.title("China Cases")
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Cases')
plt.show()
```

```python
import matplotlib.path as mpath
```

```python
plt.style.use('fivethirtyeight')

star = mpath.Path.unit_regular_star(6)
circle = mpath.Path.unit_circle()
# concatenate the circle with an internal cutout of the star
verts = np.concatenate([circle.vertices, star.vertices[::-1, ...]])
codes = np.concatenate([circle.codes, star.codes])
cut_star = mpath.Path(verts, codes)

x = Top_Countries.iloc[0,2:9]
fig, ax = plt.subplots()
plt.xticks(rotation=45)

ax.plot(x,marker=cut_star,markersize=15)
ax.set_title("China - Date Wise Chart")

plt.show()
```

## Continent Wise  [Third-party dataset]

```python
asia = pd.read_csv("as_countries.csv")
eu = pd.read_csv("eu_countries.csv")
af = pd.read_csv("af_countries.csv")
na = pd.read_csv("na_countries.csv")
sa = pd.read_csv("sa_countries.csv")
```

```python
as_countries = asia['name'].values.tolist()
eu_countries = eu['name'].values.tolist()
af_countries = af['name'].values.tolist()
na_countries = na['name'].values.tolist()
sa_countries = sa['name'].values.tolist()
date = "3/10/20"
```

```python
cases = dt[date]
countries = dt["Country/Region"]
```

```python
# counter for Countries

ctr = 0
as_cases = 0
eu_cases = 0
af_cases = 0
na_cases = 0
sa_cases = 0
for i in countries:
    if i in as_countries:
        as_cases+=cases[ctr]
    if i in eu_countries:
        eu_cases+=cases[ctr]
    if i in af_countries:
        af_cases+=cases[ctr]
    if i in na_countries:
        na_cases+=cases[ctr]
    if i in sa_countries:
        sa_cases+=cases[ctr]
    ctr+=1
```

```python
cases_n = [as_cases, eu_cases, af_cases, na_cases, sa_cases]
countries_n = ['Asia', 'Europe', 'Africa', "North America", "South America"]
```

```python
plt.rcdefaults()
plt.barh(countries_n, cases_n, align='center')
plt.title('Continent Graph for COVID')
plt.show()
```
